```python
import numpy as np
import pandas as pd
import os
```

```python
NFL_Data = pd.read_csv('thirdDown.csv')
```

```python
NFL_Data.head()
```

```python
NFL_Data = NFL_Data.drop(columns = ['Unnamed: 0', 'down'])
```

```python
#run = 1, pass =2
NFL_Data=NFL_Data.replace(['run', 'pass'], 
                     [1, 2])
NFL_Data.head()
```

```python
Run_Data = NFL_Data.loc[NFL_Data['play_type'] == 1]
```

```python
Run_Data.head()
```

```python
Run_Data = Run_Data.drop(columns = ['play_id', 'game_id','pass_length', 'pass_length.1'])
```

```python
Run_Data.head()
```

```python
Pass_Data = NFL_Data.loc[NFL_Data['play_type'] == 2]
```

```python
Pass_Data = Pass_Data.drop(columns = ['play_id', 'game_id', 'run_location'])
```

```python
# 1 =  right, 2 = middle, 3 = left
Pass_Data = Pass_Data.replace(['right', 'middle', 'left'], 
                     [1, 2, 3])
```

```python
Pass_Data.head()
```

```python
Pass_Data= Pass_Data.dropna(subset = ['pass_length'])
```

```python
Pass_Data = Pass_Data.replace(['deep', 'short'],
                 [1, 2])
```

```python
Pass_Data.head()
```

```python
Pass_Data = Pass_Data.drop(columns = ['pass_length.1'])
```

```python
Run_Data = Run_Data.dropna(subset = ['run_location'])
```

```python
Run_Data = Run_Data.replace(['right', 'middle', 'left'],
                           [1, 2, 3])
```

```python
Run_Data = Run_Data.drop(columns = ['play_type'])
```

```python
Run_Data.dtypes
```

```python
Pass_Data = Pass_Data.drop(columns = ['play_type'])
Pass_Data.head()
```

```python
Run_Data.head()
```

```python
Success_Data = NFL_Data.loc[NFL_Data['third_down_converted'] == 1]
```

```python
Success_Data.head()
```

```python
Success_Data = Success_Data.drop(columns = ['play_id', 'game_id'])
```

```python
Success_Data = Success_Data.replace(['deep', 'short'],
                 [1, 2])
```

```python
Success_Data = Success_Data.replace(['right', 'middle', 'left'], 
                     [1, 2, 3])
```

```python
Success_Data = Success_Data.drop(columns = ['third_down_converted'])
```

```python
Success_Data = Success_Data.drop(columns = ['pass_length.1'])
```

```python
Success_Data.dtypes
Success_Data.head()
```
